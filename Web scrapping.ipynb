```python
ul = soup.find('ol', {'class': 'products list items product-items'}) 

if ul is not None:
    items = ul.find_all('li', {'class': 'item product product-item'})
    leng=len(items)
    print(leng)
else:
    print('Could not find the unordered list element')

if items:
    first_item = items[0]  
else:
    print('No items found')
    
```

```python
info=[]
for item in items:
  record={}
  record['Item name']=(item.find('strong',{'class':"product name product-item-name"}).text)
#   record['Final Price']=str(item.find('span',{'class':"price-wrapper price-including-tax"}).text)
#   record['Including VAT']=(item.find('span',{'class':"price-wrapper price-excluding-tax"}).text)

  info.append(record)
```
